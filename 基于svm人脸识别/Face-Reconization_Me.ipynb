# X 训练集  y 预测集

```python
X = faces['data']
y = faces['target']
target_names = faces['target_names']
```

```python
X
```

```python
X.shape
np.cumsum(X)/np.sum(X) > 0.9
```

# PCA降维
计算是数据的特征向量,去除权重大于90%

```python
pca = PCA(n_components=True)
pca.fit(X)
```

```python
X_pca = pca.transform(X)
X_pca.shape
```

```python
X_pca.mean(axis = 0)
```

```python
X_pca.std(axis = 0) 
```

# train_test_split（）是sklearn.cross_validation模块中用来随机划分训练集和测试集，

```python
from sklearn.model_selection import train_test_split
# train_test_split（）是sklearn.cross_validation模块中用来随机划分训练集和测试集，
```

```python
X_train,X_test,X_train_pca,X_test_pca,y_train,y_test = train_test_split(X,X_pca,y,test_size = 0.2)
```

# 指定超参数 使用交叉验证

```python
C = [0.001,0.1,1,5,10]
gamma = [0.0001,0.0002,0.0033,0.01,0.1]
svc = SVC()
gcv = GridSearchCV(svc,param_grid={'C':C,'gamma':gamma})
```

```python
%time gcv.fit(X_train_pca,y_train)
```

```python
gcv.best_params_
```

```python
gcv.best_score_
```

```python
y_ = gcv.predict(X_test_pca)
y_
```

```python
gcv.score(X_test_pca,y_test)
```
